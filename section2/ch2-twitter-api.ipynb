```python
# Imports
import os
import tweepy
from tweepy import OAuthHandler

# Globals (GETTING FROM ENV VARIABLES DEFINED)
# replace these with your keys and secrets.
CONSUMER_KEY = os.getenv('CONSUMER_KEY')
CONSUMER_SECRET = os.getenv('CONSUMER_SECRET')
ACCESS_TOKEN = os.getenv('ACCESS_TOKEN')
ACCESS_SECRET = os.getenv('ACCESS_SECRET')

# Set connection
auth = OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)
query = tweepy.API(auth)
```

```python
screen_names = ['PyConAU']
users = query.lookup_users(screen_name=screen_names)
for user in users:
    print(user._json)
```

```python
search_term = "natural language processing"
users = query.search_users(search_term)
for user in users:
    print(user._json)
```

To get followers or friends of a known screen name or Twitter ID,

```python
screen_name = "PyConAU"
followers = query.get_followers(screen_name=screen_name)
for fol in followers:
   	print(fol._json)
friends = query.get_friends(screen_name=screen_name)
for fr in friends:
    print(fr._json)
```

```python
# To get only IDs rather than detailed data 
# for each follower/friend,
# the below can be used instead
follower_ids = query.get_follower_ids(screen_name=screen_name)
print(follower_ids)
friend_ids = query.get_friend_ids(screen_name=screen_name)
print(friend_ids)
```

To iterate through result pages, use tweepy.Cursor

```python
# To account for rate limits
query = tweepy.API(
    auth, 
    wait_on_rate_limit=True
) # note this will cause the code to run for longer due to auto 15-min sleeps.
```

```python
screen_name = "PyConAU"
cursor = tweepy.Cursor(
    query.get_followers, screen_name=screen_name, count=200
).pages()

followers = []
for i, page in enumerate(cursor):
    followers += [itm._json for itm in page]
print("No. of follower details", len(followers))
```

```python
screen_name = "PyConAU"
alltweets = []
# make initial request for most recent tweets 
# (200 is the maximum allowed count)
new_tweets = query.user_timeline(
    screen_name=screen_name,
    count=200
)
# save most recent tweets
alltweets.extend(new_tweets)

# To get more tweets,
# save the id of the oldest tweet less one
oldest = alltweets[-1].id - 1
# Grab tweets until there are no tweets left to fetch
while len(new_tweets) > 0:
    # all subsiquent requests 
    # use the max_id param to prevent duplicates
    new_tweets = query.user_timeline(
        screen_name=screen_name,
        count=200,
        max_id=oldest
    )
    # save most recent tweets
    alltweets.extend(new_tweets)
    # update the id of the oldest tweet less one
    oldest = alltweets[-1].id - 1
    print("{} tweets downloaded\n".format(len(alltweets)))

# transform the tweepy tweets into a 2D array
all_tweets = [tweet._json for tweet in alltweets]

# Print all_tweets
print("Total count", len(all_tweets))
```
